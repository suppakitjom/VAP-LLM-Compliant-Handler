```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_excel('./Accusation Mock.xlsx',sheet_name='Sheet1')
data2 = pd.read_excel('./Mock2.xlsx',sheet_name='Sheet1')
```

```python
data1.head()
```

```python
data2.head()
```

```python
y_true = pd.concat([data1['ฐานความผิด'],data2['ประเภท']],axis=0,ignore_index=True).apply(lambda x: x.strip())
y_pred = pd.concat([data1['Category'],data2['Category']],axis=0,ignore_index=True).apply(lambda x: x.strip())
```

```python
y_true
```

```python
y_pred
```

```python
# plot confusion matrix of ytrue and ypred 
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
